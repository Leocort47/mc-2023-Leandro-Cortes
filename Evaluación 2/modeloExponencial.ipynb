```python
import numpy as np
from scipy . optimize import curve_fit
import matplotlib . pyplot as plt

x = np . array ([1, 2, 3, 4, 5, 6, 7, 8])
y = np . array ([3.8, 4, 4.2, 4.6, 4.9, 5.3, 5.7, 6.1])

xs = np . linspace (0 ,10 ,10)

def f (x ,a , b ) :
    return a * np . exp ( - b * x )

param , _ = curve_fit (f ,x ,y , p0 =[0.1 , -0.0001])
print ( param )

fig = plt . figure ()
plt.scatter (x , y , label ='Datos ')
plt.plot ( xs , f ( xs ,* param ) , ':r', label ='Regresión exponencial ')
plt.xlabel ('')
plt.ylabel (' ')
plt.legend ()
plt.title ('modelo exponencial ')
plt.grid ()
plt.show ()
#fig.savefig (" regexp . pdf", bbox_inches ='tight ')
```
