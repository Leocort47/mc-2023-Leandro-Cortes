```python
#(30%) Halle los polinomios de interpolación de Lagrange de grado 1, 2, 3 y 4 que mejor sirvan
#para estimar el valor de f(4,75) tomando como base los siguientes puntos. Estime además el
#valor de f(4,75) para cada uno de los grados. Incluya las 4 gráficas de los polinomios.

import numpy as np
import matplotlib . pyplot as plt
from scipy import interpolate

x = np . array ([1 ,3 ,3.5 ,3.8 ,4.5])
y = np . array ([2.8 ,3 ,3.5 ,3.8 ,4.5])
n = x . size
xi =3
yi =0
# Calcula los factores de Lagrange y hace la suma
for i in range (0 , n ) :
    producto = y [ i ]
for j in range (0 , n ) :
    if i != j :
        producto = producto * ( xi - x [ j ]) /( x [ i ] - x [ j ]) ;
        yi = yi + producto

print ( yi )
f = interpolate . lagrange (x , y ) # usando la funcion de Lagrange de ←-

print ( f ( xi ) )

plt.plot (x ,y ,'o')
plt.plot ( xi , yi ,'sr ')
plt.text ( xi +0.1 , yi , ' interpolacion ' + str( yi ) )
plt.title ('Profundidad ')
plt.legend ([ 'Datos ',' Interpolacion '])
plt.grid ( True )
plt.show ()
```
