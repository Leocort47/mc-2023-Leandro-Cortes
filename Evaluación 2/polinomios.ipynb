```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de los puntos de muestra
x = np.array([1, 3, 5, 7, 9])
y = np.array([2.8, 3, 3.5, 3.8, 4.5])

# Función para calcular los polinomios de interpolación de Lagrange
def lagrange_interpolation(x, y, degree, x_val):
    n = len(x)
    polynomial = 0
    
    for i in range(n):
        numerator = 1
        denominator = 1
        for j in range(n):
            if i != j:
                numerator *= (x_val - x[j])
                denominator *= (x[i] - x[j])
        
        polynomial += (numerator / denominator) * y[i]
        
        if degree == i + 1:
            break
    
    return polynomial

# Valor de x para estimar
x_val = 4.75

# Grado 1 (Polinomio lineal)
poly_degree1 = lagrange_interpolation(x, y, 1, x_val)

# Grado 2 (Polinomio cuadrático)
poly_degree2 = lagrange_interpolation(x, y, 2, x_val)

# Grado 3 (Polinomio cúbico)
poly_degree3 = lagrange_interpolation(x, y, 3, x_val)

# Grado 4 (Polinomio de grado 4)
poly_degree4 = lagrange_interpolation(x, y, 4, x_val)

# Estimación de f(4.75)
f_x_val = lagrange_interpolation(x, y, len(x) - 1, x_val)

print(f"Estimación de f(4.75) para cada grado:")
print(f"Grado 1: {poly_degree1}")
print(f"Grado 2: {poly_degree2}")
print(f"Grado 3: {poly_degree3}")
print(f"Grado 4: {poly_degree4}")
print(f"Estimación final de f(4.75): {f_x_val}")

# Gráficas de los polinomios
x_range = np.linspace(min(x), max(x), 100)
y_range1 = lagrange_interpolation(x, y, 1, x_range)
y_range2 = lagrange_interpolation(x, y, 2, x_range)
y_range3 = lagrange_interpolation(x, y, 3, x_range)
y_range4 = lagrange_interpolation(x, y, 4, x_range)

plt.plot(x, y, 'ro', label='Puntos de muestra')
plt.plot(x_range, y_range1, label='Grado 1')
plt.plot(x_range, y_range2, label='Grado 2')
plt.plot(x_range, y_range3, label='Grado 3')
plt.plot(x_range, y_range4, label='Grado 4')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.title('Polinomios de Interpolación de Lagrange')
plt.show()
```
